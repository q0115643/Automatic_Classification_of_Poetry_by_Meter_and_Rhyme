```python
import torch
from torch.autograd import Variable
from torch.utils.data import Dataset, DataLoader
import csv
import string
import json

all_letters = string.ascii_letters + string.punctuation + " "
n_letters = len(all_letters)
in_length = 80
max_lines = 20

def letterToTensor(letter):
    tensor = torch.zeros(1, n_letters)
    tensor[0][all_letters.find(letter)] = 1
    return tensor

def lineToTensor(line):
    tensor = torch.zeros(in_length, n_letters)
    for li, letter in enumerate(line):
        tensor[li][all_letters.find(letter)] = 1
    return tensor

def poemToTensor(lines):
    a = [lineToTensor(line) for line in lines]
    while len(a) < max_lines:
        a.append(torch.zeros(in_length, n_letters))
    return torch.cat([torch.unsqueeze(x, 0) for x in a])

def labelToTensor(line):
    tensor = torch.zeros(1, 2)
    if line is 'metrical':
        tensor[0][0] = 1
    else:
        tensor[0][1] = 1
    return tensor

class MyDataset(Dataset):
    def __init__(self):
        x = []
        y = []
        with open('poetry.csv', newline='') as myFile:  
            reader = csv.reader(myFile)
            for row in reader:
                x.append(poemToTensor(json.loads(row[0])[:max_lines]))
                y.append(labelToTensor(row[1]))
        self.len = len(x)
        self.x_data = torch.cat([torch.unsqueeze(a, 0) for a in x])
        self.y_data = torch.cat([torch.unsqueeze(a, 0) for a in y])

    def __getitem__(self, index):
        return self.x_data[index], self.y_data[index]

    def __len__(self):
        return self.len


dataset = MyDataset()
train_loader = DataLoader(dataset=dataset,
                          batch_size=32,
                          shuffle=True)
```

```python
for i, data in enumerate(train_loader, 0):
    inputs, labels = data
    print(inputs.shape, labels.shape)
```
